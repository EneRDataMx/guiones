```python
data = pd.read_csv(f)
ind = data
data.set_index('A単o', inplace= True)
data
```

```python
fig, ax = plt.subplots(constrained_layout=True)

ax.plot(ind["Porcentaje ER"], '-o', color = '#96c120',
        label="Evoluci坦n ER")

ax.plot(ind["Porcentaje EL"],'-o', color = '#8990a2', 
        label="Evoluci坦n EL")

ax.set_ylabel("Porcentaje")
ax.set_xlabel("A単o")
ax.set_ylim(10,35)
ax.legend()
plt.savefig("Evolucion_EL_ER.jpg")
```

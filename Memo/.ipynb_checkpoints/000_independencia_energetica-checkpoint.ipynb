```python
import altair as alt
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
f   = '../data/independencia.csv'
ind = pd.read_csv(f,index_col=0,parse_dates=True)
ind['A침o'] = ind.index.year
ind.info()
```

```python
fig, ax = plt.subplots()

kwargs = {"lw":2,"ms":10,"alpha":1,"mfc":"white"}

ax.plot(ind["Producci칩n"], '-o',
        label="Producci칩n Primaria",**kwargs)

ax.plot(ind["Oferta Interna Bruta"], '-*',
        label="Oferta Interna Bruta",**kwargs)

ax.set_ylabel("Energ칤a en PetaJoules")
ax.set_xlabel("A침o")
ax.set_ylim(6000,10000)
ax.legend()

ax.spines[['top','right']].set_visible(False)
```

```python
fig, ax = plt.subplots()

ax.axhline(y=1,c="k",alpha=0.3)
ax.plot(ind.Factor, '-o', ms=10, lw=2, alpha=1, mfc='white')

ax.set_xlim(ind.index[0]-pd.Timedelta("1Y"),ind.index[-1]+pd.Timedelta("1Y"))
ax.set_ylim(0,1.2)

ax.set_ylabel("Factor de Independencia")
ax.set_xlabel("A침o")

ax.spines[['top','right']].set_visible(False)
```

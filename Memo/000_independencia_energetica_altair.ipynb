```python
f   = '../data/independencia.csv'
ind = pd.read_csv(f,index_col=0,parse_dates=True)
ind['AÃ±o'] = ind.index.year.astype(str)
# ind.info()
```

```python
fig, ax = plt.subplots(constrained_layout=True)


ax.plot(ind["ProducciÃ³n"], '-o',
        label="ProducciÃ³n Primaria")

ax.plot(ind["Oferta Interna Bruta"], '-*',
        label="Oferta Interna Bruta")

ax.axvspan(parse("2011-01-01"),parse("2013-12-31"),alpha=0.1,color="g")
ax.axvspan(parse("2013-12-31"),parse("2020-12-31"),alpha=0.1,color="r")

ax.set_ylabel("EnergÃ­a [$PJ$]")
ax.set_xlabel("AÃ±o")
ax.set_ylim(6000,10000)
ax.legend()


ax.spines[['top','right']].set_visible(False)
```

```python

consumo = alt.Chart(ind).mark_line(color='green').encode(
    alt.X('AÃ±o:T', axis=alt.Axis(format='%y')),
    alt.Y('ProducciÃ³n',scale=alt.Scale(zero=False),title="EnergÃ­a [PJ]")
).properties(
    width=300,
    height=300
)
point_consumo = consumo.mark_point(color='green')


produccion = alt.Chart(ind).mark_line(color='red').encode(
    alt.X('AÃ±o:T', axis=alt.Axis(format='%Y')),
    alt.Y('Oferta Interna Bruta',scale=alt.Scale(zero=False))
)
point_produccion = produccion.mark_point(color='red')


consumo + point_consumo + produccion + point_produccion

```
